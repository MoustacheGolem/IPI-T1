```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#imagem original:
img = cv2.imread("Image1.pgm")

#mostrar:
plt.figure(figsize=(15,10))
plt.axis("off")
plt.imshow(img)
```

```python
#*** QUESTAO 2.1 ***
#filtro laplaciano:
filtroLaplaciano8 = np.array([[1, 1,1],
                             [1,-8,1],
                             [1, 1,1]])

imagemLaplaciana = cv2.filter2D(src=img,ddepth=-1,kernel=filtroLaplaciano8) #aplicar filtro
resultado2_1 = img + -1*imagemLaplaciana        # somar imagem filtrada(detalhes) a original

#mostrar:
plt.figure(figsize=(15,10))
plt.axis("off")
plt.imshow(resultado2_1)
```

```python
#*** QUESTAO 2.2 ***
#  suavizao Gaussiana:
standard_deviation = 0.5
blurred_img = cv2.GaussianBlur(img,(3,3),standard_deviation)

# filtro laplaciano:
filtroLaplaciano4 = np.array([[0, 1,0],
                             [1,-4,1],
                             [0, 1,0]])
                            
imagemLaplaciana = cv2.filter2D(src=blurred_img,ddepth=-1,kernel=filtroLaplaciano4)


resultado2_2 = blurred_img + -1*imagemLaplaciana

# mostrar
plt.figure(figsize=(15,10))
plt.axis("off")
plt.imshow(resultado2_2)
```

```python
#*** QUESTAO 2.3 ***
#  suavizao Gaussiana:
standard_deviation = 1
filtroLaplaciano = np.array([[0, 1,0],
                             [1,-4,1],
                             [0, 1,0]])
                            

blurred_img = cv2.GaussianBlur(img,(3,3),standard_deviation)


# filtro laplaciano:
imagemLaplaciana = cv2.filter2D(src=blurred_img,ddepth=-1,kernel=filtroLaplaciano)


resultado2_3 = blurred_img + -1*imagemLaplaciana

#mostrar:
plt.figure(figsize=(15,10))
plt.axis("off")
plt.imshow(blurred_img)
```
